```python
import matplotlib.pylab as plt
import numpy as np
```

```python
def step(xi, yi, lx, ux, ly, uy):

    r = np_rng.integers(low=1, high=5)
    if r == 1:
        # go east
        if xi < ux:
            xi = xi + 1
    elif r == 2:
        # go west
        if xi > lx:
            xi = xi - 1
    elif r == 3:
        # go north
        if yi < uy:
            yi = yi + 1
    else:
        # go south
        if yi > ly:
            yi = yi - 1
    return xi, yi
```

```python
# --Main Program

# Set up random number generator
np_rng = np.random.default_rng(seed=31415)
for i in range(1,1000):
    random_float = np_rng.random()

# define the grid
lx = -40
ux = 40
ly = -40
uy = 40

# set up the simulation
N = 1000
xi = 0
yi = 0
position_array = np.zeros((N + 1, 2))

# execute random walk
for i in range(1, N + 1):
    xi, yi = step(xi, yi, lx, ux, ly, uy)
    position_array[i, 0] = xi
    position_array[i, 1] = yi

x_array = position_array[:, 0]
y_array = position_array[:, 1]


# plot distance vs time instead of the 2D array

def distance_from_origin(x, y) -> float:
    a = x
    b = y

    a *= a
    b *= b

    c = (a+b)**0.5

    return c

distance:list[float] = [distance_from_origin(x_array[i], y_array[i]) for i in range(N)]

time:list[int] = list(range(1, N+1))

# create plot
x0 = x_array[0]  # initial x
y0 = y_array[0]  # initial y
xf = x_array[-1]  # final x
yf = y_array[-1]  # final y

print(len(distance))
plt.scatter(x=time, y=distance)

plt.xlabel("Time")
plt.ylabel("Distance from Origin")

plt.savefig('walk_distance.png', dpi=300)
plt.show()
```
