### Program: Zero Integer Checker
#### Author: Johang Hernandez
#### Version: 1.30.2026.1
#### Summary: This program will ask for two inputs, then tell the user if both integers are zero or not.

```python
# Ask user for two integers
i1 = input('type in an integer ')
i1 = int(i1)
i2 = input('type in another integer ')
i2 = int(i2)
if i1 == 0 and i2 == 0:
    print('Both integers are zero')
else:
    print('Both integers are not zero')
```
