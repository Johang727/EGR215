> Summary: Temperature Conversion Program with Input error checking of temperatures. This version accepts a user-defined temperature in Celsius, Fahrenheit, or Kelvin and converts the temperature to a user selected scale.

```python
# Display program greeting
print('Welcome to the Temperature Scale Conversion Program!')
print('This program will request that the user submit a temperture.')
print('Next, it requests the converted scale.')
print('Finally, it prints out the converted temperature.')
submission_is_incorrect = True

# Gets scale to convert from
scale_request = "'C' for Celsius, 'F' for Fahrenheit, 'K' for Kelvin, 'R' for Rankine."

valid_scales:list[str] = ["C", "F", "K", "R"]

def conv_fahr_to_cel(fahr:float) -> float:
    return (fahr - 32.0) * 5.0/9.0
def conv_cel_to_fahr(cel:float) -> float:
    return (cel * 9.0/5.0) + 32.0





while submission_is_incorrect:
    submitted_temperature = float(input('Submit a temperature to convert: '))
    print('Specify the scale of your submitted temperature: ')
    submitted_scale = input(scale_request)
    if ((submitted_scale == 'C') and (submitted_temperature >= -273.15)):
        submission_is_incorrect = False
    elif ((submitted_scale == 'F') and (submitted_temperature >= -459.67)):
        submission_is_incorrect = False
    elif ((submitted_scale == 'K') and (submitted_temperature >= 0)):
        submission_is_incorrect = False
    elif submitted_scale == "R" and submitted_temperature >= 0:
        submission_is_incorrect = False
    else:
        print('Incorrect submitted temperature. Try again.')

# remove current scale from valid list
valid_scales.remove(submitted_scale)

# Gets scale to convert to
print('What scale should be used for the converted temperature?')
converted_scale = input(scale_request)
submission_is_incorrect = True

while submission_is_incorrect:
    if converted_scale in valid_scales:
        submission_is_incorrect = False
    else:
        print('Incorrect scale to convert to. Try again.')
        converted_scale = input(scale_request)

# Celsius
if submitted_scale == 'C':
    if converted_scale == 'F':
        # convert Celsius to Fahrenheit
        converted_temperature = conv_cel_to_fahr(submitted_temperature)
    elif converted_scale == "R":
        # convert Celsius to Rankine
        converted_temperature = conv_cel_to_fahr(submitted_temperature) + 459.67
    else:
        # convert Celsius to Kelvin
        converted_temperature = submitted_temperature + 273.15

# Fahrenheit
elif submitted_scale == 'F':
    if converted_scale == 'C':
        # convert Fahrenheit to Celsius
        converted_temperature = conv_fahr_to_cel(submitted_temperature)
    elif converted_scale == "R":
        converted_temperature = submitted_temperature + 459.67
    else:
        # convert Fahrenheit to Kelvin
        converted_temperature = conv_fahr_to_cel(submitted_temperature + 491.67)

# Rankine
elif submitted_scale == "R":
    if converted_scale == "F":
        converted_temperature = submitted_temperature - 459.67
    elif converted_scale == "C":
        converted_temperature = conv_fahr_to_cel(submitted_temperature - 459.67)
    else: # Kelvin
        converted_temperature = submitted_temperature * (5.0/9.0)

# Kelvin
else:
    if converted_scale == 'C':
        # convert Kelvin to Celsius
        converted_temperature = submitted_temperature - 273.15
    elif converted_scale == "R":
        converted_temperature = submitted_temperature * (9.0/5.0)
    else:
        # convert Kelvin to Fahrenheit
        converted_temperature = conv_cel_to_fahr(submitted_temperature - 273.15)

# print out the result
s1 = format(submitted_temperature, '.2f')
s2 = format(converted_temperature, '.2f')
print(s1, submitted_scale, ' is ', s2, converted_scale)
```
