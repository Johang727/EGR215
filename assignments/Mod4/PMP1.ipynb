```python
"""
Program: Count Odds
Author: C.D. Wentworth
Version: 2.5.2020.1
Summary: This program creates a list of random numbers and
         counts how many odd numbers are in the list.
         
"""
import random

# make a list of random numbers
lst = []
for i in range(10):
    lst.append(random.randint(0, 1000))

# make another list
lst2 = []
for i in range(10):
    lst2.append(random.randint(0, 1000))


# sum odd numbers in the list
odd_sum:int = 0
for e in lst:
   if e % 2 != 0:
       odd_sum += e

# sum odd numbers in the list
even_sum:int = 0
for e in lst2:
   if e % 2 == 0:
       even_sum += e

print(f"""
The sum of the odd numbers in the first list is {odd_sum}.
The sum of the even numbers in the second list is {even_sum}.
The sum of both numbers is {odd_sum + even_sum}.
""")
```
