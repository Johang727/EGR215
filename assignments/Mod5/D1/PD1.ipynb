<p>Summary: This program plots two graphs:</p>
<ul>
<li><i>f1(t) = t - t^2</i></li>
<li><i>f2(t) = t^2 - t^3</t></li>
</ul>


```python
# managing imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate a range of values from 0 to 2
t = np.linspace(0, 2)

# generates f values for both graphs
f1 = t - (t**2)
f2 = t**2 - (t**3)
```

```python
# sets y axis limits rateher than using defaults
plt.gca().set_ylim((-4.5, 2.5))

# plots both f functions
plt.plot(t, f1, color="b", label = "f1", linestyle="-")
plt.plot(t, f2, color="b", label = "f2", linestyle="--")

# set labels
plt.xlabel("t")
plt.ylabel("f")

# display legend
plt.legend()

# show graph
plt.show()
```
