```python
print("Welcome to the temperature conversion program!")

# get temperature in Fahrenheit
TC:float = float(input('Enter degrees Celsius: '))

# calc degrees Celsius
TF:float = (TC*1.8) + 32

# output degrees Celsius
print(f"{TC:.2f} degrees Celsius equals {TF:.2f} degrees Fahrenheit!")
```

```python
"""
Program: Temperature Conversion Program (Fahrenheit to Celsius)
Author: 
Version: 1.25.2020.1
Summary: This program will convert a temperature entered in Fahrenheit
         to the equivalent temperature in Celsius

"""

# get temperature in Fahrenheit
TF = float(input('Enter degrees Fahrenheit: '))

# calc degrees Celsius
TC = (TF - 32) * 5 / 9

# output degrees Celsius
print(TF, 'degrees Fahrenheit equals',
       format(TC, '.1f'), 'degrees Celsius')
```
