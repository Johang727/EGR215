<p>Summary: This program produces a table of equivalent temperatures in the Kelvin, Celsius, Fahrenheit, and Rankine scales from 0 Kelvin to 375 Kelvin.</p>
<ul>

```python
# manage imports
import numpy as np
```

```python
# make functions

# kelvin to celsius
def KtoC(temp:float) -> float:
    return (temp - 273.15)

# kelvin to fahrenheit
def KtoF(temp:float) -> float:
    return (KtoC(temp)*1.8) + 32

# kelvin to rankine
def KtoR(temp:float) -> float:
    return temp*1.8
```

```python
# make empty lists
kelvin_temps = np.empty(76)
celsius_temps = np.empty(76)
fahrenheit_temps = np.empty(76)
rankine_temps = np.empty(76)

# get temperature conversions and populate lists
for i, temp in enumerate(range(0, 380, 5)):
    #print(i, temp)
    kelvin_temps[i] = temp
    celsius_temps[i] = KtoC(temp)
    fahrenheit_temps[i] = KtoF(temp)
    rankine_temps[i] = KtoR(temp)

# combine lists
temperatures = np.column_stack((kelvin_temps, celsius_temps, fahrenheit_temps, rankine_temps))

np.savetxt("temps.txt",
           temperatures,
           fmt='%.2f',
           header="Kelvin\tCelsius\tFahrenheit\tRankine",
           delimiter="\t")
```
